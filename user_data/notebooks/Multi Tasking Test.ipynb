```python
import multitasking
import requests
import json
from time import perf_counter

multitasking.set_max_threads(10)
start = perf_counter()

class DownloadStuff:

    def downloader(self, id):
        start = perf_counter()
        print(f"Starting Task# {id}")
        r = requests.get("https://httpbin.org/get").text
        print(f"Finished Task# {id}. Time taken {(perf_counter() - start):.2f}s")
        return r

    def save_results(self, r):
        with open("response.json", "a") as f:
            f.write(r)
            
    @multitasking.task
    def wrapper(self):
        self.run()

    def run(self):
        for i in range(0, 50):
            r = self.downloader(i)
            self.save_results(r)

        multitasking.wait_for_tasks()
        print(f"Total time taken {(perf_counter() - start):.2f}s")
```

```python
d = DownloadStuff()
```

```python
d.run()
```
