```python
df = pd.read_sql_table("feature_importance", "sqlite:///litmus.sqlite")
df
```

```python
df["selected"].mean()
```

```python
df_stats = df.groupby(["pair", "feature_name"]).agg(
    mean_imp=('Importances', np.mean), 
    mean_rank=('rank', np.mean),
    mean_selected=('selected', np.mean),
    std_imp=('Importances', np.std),
    std_rank=('rank', np.std))
df_stats
```

```python
df_stats.sort_values(["pair", "mean_rank"], ascending=True).tail(50)
```
